## Вступ до кластеризації

**Кластеризація** &mdash; це метод аналізу даних, який розділяє дані на групи (кластери) так, щоб об’єкти в одній групі були схожі між собою, а між різними групами відрізнялися.

Кластеризація належить до підходів **навчання без учителя**, тобто працює з даними, які не мають наперед заданих міток або правильних відповідей. У такому випадку алгоритм самостійно аналізує структуру даних, виявляє приховані закономірності та на їх основі формує групи об’єктів (кластери). Для цього застосовуються різні алгоритми, які оцінюють схожість між об’єктами та об’єднують їх у кластери.

У науці про дані кластеризація використовується для аналізу уподобань користувачів або дослідження будь-яких немаркованих даних. Кластеризація допомагає навести лад у хаосі, коли даних багато, але вони не структуровані.

**Наприклад, кластеризацію застосовують:**


*   коли потрібно розбити дані на групи без готових міток;
*   коли потрібно знайти схожі об’єкти;
*   коли потрібно виявити приховані закономірності в даних.

## Принципи кластеризації

### 1.** Кластеризація на основі центроїдів**

Алгоритми цього типу формують кластери на основі **відстані об’єктів до центрів (центроїдів)**.

📌 **Ключова ідея:**
кожен об’єкт належить до того кластера, центр якого знаходиться найближче.


Існує багато способів вимірювання відстані між об’єктами. Найпоширеніші з них:

- **Евклідова відстань** — це “звичайна” відстань між точками (як по прямій, яку можна виміряти лінійкою).  
  Добре працює для простих задач, але може бути менш ефективною у багатовимірних просторах.

- **Мангеттенська відстань** — це відстань, яку потрібно пройти, рухаючись лише вздовж осей (по “сітці”).  
  Назва походить від планування вулиць у місті Мангеттен: рух можливий тільки вздовж кварталів.

👉 У багатовимірних задачах мангеттенська метрика часто дає кращі результати.


**Метод k-середніх (k-means)** — це один із найпопулярніших алгоритмів кластеризації на основі центроїдів. Його ідея полягає в тому, що кластери формуються залежно від відстані між точками даних і центрами кластерів.

Перед початком роботи необхідно задати кількість кластерів `k`. Після цього алгоритм випадковим чином розміщує початкові центри кластерів і починає ітеративний процес. На кожному кроці кожна точка відноситься до найближчого центру, після чого центри перераховуються як середнє значення всіх точок, що належать до відповідного кластера. Цей процес повторюється доти, поки положення центрів не стабілізується.

Алгоритм є досить простим і ефективним для невеликих наборів даних, однак має певні обмеження. Зокрема, він вимагає заздалегідь визначити кількість кластерів, є чутливим до шуму та викидів і може давати різні результати через випадкову ініціалізацію центрів. Саме тому на практиці його часто запускають кілька разів і обирають найкращий результат.

![k-means clustering](https://robot-site-static.fra1.digitaloceanspaces.com/cc/ckeditor/blog/93-obedinyay-i-vlastvuy-algoritmy-klasterizacii/kmeans.gif)

Анімація демонструє, як центроїди поступово зміщуються, а точки “перерозподіляються” між кластерами до досягнення стабільного результату.



> Окрім методу k-середніх, існують й інші алгоритми кластеризації, побудовані на основі центроїдів, зокрема **K-medoids, CLARANS, Fuzzy C-means** та **generalized K-harmonic means**. Вони розвивають ідею k-means, але по-різному підходять до вибору центрів кластерів або способу віднесення об’єктів до них, що дозволяє зменшити чутливість до шуму або підвищити якість кластеризації в складніших випадках.

### 2. **Кластеризація на основі розподілу**

Кластеризація цього типу базується на припущенні, що дані утворені з кількох **ймовірнісних розподілів** (наприклад, гауссового або біноміального), які задаються заздалегідь.

На відміну від кластеризації на основі центроїдів, об’єкти відносять до кластерів не за відстанню до центру, а за принципом **максимальної правдоподібності** — тобто до того кластера, де ймовірність належності є найбільшою.

Кожен кластер при цьому описується не лише своїм центром, а й параметрами розподілу (наприклад, середнім значенням і дисперсією), що дозволяє більш точно відображати структуру даних.

Ключовим прикладом кластеризації на основі розподілів є **Gaussian Mixture Models (GMM)** — модель гаусової суміші.

Вона базується на використанні **гауссового (нормального) розподілу**, який описує, як імовірності розподіляються навколо середнього значення.

GMM є гнучкішою за k-means, оскільки:
- дозволяє точкам мати **ймовірнісну належність** до кількох кластерів одночасно  
- може моделювати **кластери різної форми та розміру**  

Таким чином, кожен об’єкт не жорстко прив’язаний до одного кластера, а має набір ймовірностей належності до різних кластерів.

### 3. **Кластеризація на основі центроїдів**

Алгоритми цього типу формують кластери на основі **щільності розташування точок**. Вони знаходять області, де дані зосереджені щільно, і об’єднують їх у кластери.

Такі методи дозволяють:
- виявляти кластери **довільної форми**  
- ігнорувати поодинокі точки, які вважаються **шумом**  


Найпопулярнішим алгоритмом цього типу є **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**.

Ідея алгоритму полягає в тому, що точки вважаються пов’язаними, якщо вони знаходяться достатньо близько одна до одної.

Для формування кластерів використовуються два основні параметри:
- **eps** — максимальна відстань, в межах якої точки вважаються сусідніми  
- **minPts** — мінімальна кількість точок для утворення щільної області  

Якщо точка має достатню кількість сусідів у радіусі eps, вона вважається частиною кластера. Якщо ні — може бути віднесена до шуму.


Результат роботи алгоритму сильно залежить від правильного вибору параметрів **eps** та **minPts**.

### 4. **Ієрархічна кластеризація**

Ієрархічна кластеризація будує **дерево кластерів (дендрограму)**, де менші кластери поступово об’єднуються у більші або, навпаки, розділяються.

📌 Ідея схожа на структуру файлів у комп’ютері:
папки можуть містити вкладені папки та файли, утворюючи ієрархію.


Перевагою цього підходу є те, що **кількість кластерів не потрібно задавати наперед** — її можна обрати вже після побудови дерева.

Водночас метод:
- потребує значних обчислювальних ресурсів  
- погано масштабується для великих наборів даних  
- є чутливим до шуму та викидів  


Існує два основні підходи:

- **Висхідна (агломеративна)** — починає з окремих точок і поступово об’єднує їх у більші кластери  
- **Низхідна (дивізивна)** — починає з одного великого кластера і поступово розділяє його  


**Agglomerative Hierarchical Clustering** &mdash; це найпоширений варіант ієрархічної кластеризації.

Алгоритм працює так:
- на початку кожен об’єкт є окремим кластером  
- далі кластери послідовно об’єднуються на основі їхньої схожості  
- процес триває, поки всі об’єкти не об’єднаються в один кластер  

Для цього використовується **матриця подібності (або відстаней)**.

👉 Такий підхід добре підходить для виявлення **невеликих і чітко виражених кластерів**.

## Метод k-середніх (k-means)


### Алгоритм

1. Ініціалізуйте центри кластерів, випадковим чином в межах діапазону вхідних даних, або (рекомендовано) деякими з існуючих прикладів навчання

2. До збіжності

   2.1. Призначте кожну точку даних найближчому кластеру. Відстань між точкою та центром скупчення вимірюється за допомогою евклідової відстані.

   2.2. Оновіть поточні оцінки центрів кластера, встановивши для них середнє значення всіх екземплярів, що належать до цього кластера  
   
   
### Цільова функція

Основна цільова функція намагається знайти центри кластерів таким чином, що, якщо дані розділені на відповідні кластери, відстань між точками даних та їх найближчими центрами кластера стає якомога меншою.

Враховуючи набір точок даних $ {x_1, ..., x_n} $ і додатне число $ k $, знайдіть кластери $ C_1, ..., C_k $, які мінімізують

\begin{equation}
J = \sum_{i=1}^n \, \sum_{j=1}^k \, z_{ij} \, || x_i - \mu_j ||_2
\end{equation}

Де:  
- $z_ {ij} \in \{0,1\}$ визначає, чи не належить точка даних $ x_i $ до кластеру $ C_j $
- $ \mu_j $ позначає центр кластера $ C_j $
- $ || \,\, || _2 $ позначає евклідову відстань

### Недоліки K-Means
- Кількість кластерів потрібно встановити на початку
- Результати залежать від початкових центрів кластера
- Чутливий до викидів
- Не підходить для пошуку неопуклих скупчень (кластери складної форми)
- ...



> **Опуклий кластер** — це така множина, де будь-які дві точки можна з’єднати прямою лінією, і вона повністю лежить всередині кластера.



```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.datasets import make_blobs
np.random.seed(123)

%matplotlib inline
```

## Набір даних

```python
X, y = make_blobs(centers=4, n_samples=1000)
print(f'Shape of dataset: {X.shape}')

fig = plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Dataset with 4 clusters")
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()
```

## K-Means клас

```python
class KMeans():
    def __init__(self, n_clusters=4):
        self.k = n_clusters

    def fit(self, data):
        """
        Fits the k-means model to the given dataset
        """
        n_samples, _ = data.shape
        # initialize cluster centers
        self.centers = np.array(random.sample(list(data), self.k))
        self.initial_centers = np.copy(self.centers)

        # We will keep track of whether the assignment of data points
        # to the clusters has changed. If it stops changing, we are
        # done fitting the model
        old_assigns = None
        n_iters = 0

        while True:
            new_assigns = [self.classify(datapoint) for datapoint in data]

            if new_assigns == old_assigns:
                print(f"Training finished after {n_iters} iterations!")
                return

            old_assigns = new_assigns
            n_iters += 1

            # recalculate centers
            for id_ in range(self.k):
                points_idx = np.where(np.array(new_assigns) == id_)
                datapoints = data[points_idx]
                self.centers[id_] = datapoints.mean(axis=0)

    def l2_distance(self, datapoint):
        dists = np.sqrt(np.sum((self.centers - datapoint)**2, axis=1))
        return dists

    def classify(self, datapoint):
        """
        Given a datapoint, compute the cluster closest to the
        datapoint. Return the cluster ID of that cluster.
        """
        dists = self.l2_distance(datapoint)
        return np.argmin(dists)

    def plot_clusters(self, data):
        plt.figure(figsize=(12,10))
        plt.title("Initial centers in black, final centers in red")
        plt.scatter(data[:, 0], data[:, 1], marker='.', c=y)
        plt.scatter(self.centers[:, 0], self.centers[:,1], c='r')
        plt.scatter(self.initial_centers[:, 0], self.initial_centers[:,1], c='k')
        plt.show()
```

## Ініціалізація та підгонка моделі

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

## Побудова початкового та кінцевого кластерного центру



```python
kmeans.plot_clusters(X)
```
